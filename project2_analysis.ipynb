```python
import pandas as pd
import numpy as np

                                            
                                        # Cleaning Ticket_Price_Cr

                                        
# Copy dataset to avoid modifying original
df_clean = df.copy()

# Step 1: Remove commas and convert to numeric
df_clean['Ticket_Price_Cr'] = (
    df_clean['Ticket_Price_Cr']
    .astype(str)                           # ensure string
    .str.replace(',', '', regex=False)     # remove commas
    .str.extract(r'(\d+\.?\d*)')[0]        # extract only numeric part
    .astype(float)                         # convert to float
)

# Step 2: Check for missing or invalid values

print("Unique count:", df_clean['Ticket_Price_Cr'].nunique())
print("Missing values:", df_clean['Ticket_Price_Cr'].isna().sum())

# Step 3: Quick sanity check
df_clean[['Ticket_Price_Cr']].head(10)
```

```python
                                    # Cleaning Unit_Size_Sqft

# Step 1: Convert to string
df_clean['Unit_Size_Sqft'] = df_clean['Unit_Size_Sqft'].astype(str)

# Step 2: Handle ranges like "1200-1500" → take average
def parse_area(value):
    # Extract numbers
    nums = [float(x) for x in value.replace(',', '').split('-') if x.strip().replace('.', '').isdigit()]
    if len(nums) == 1:   # single value
        return nums[0]
    elif len(nums) == 2: # range -> average
        return sum(nums) / 2
    else:
        return np.nan    # invalid or text

df_clean['Unit_Size_Sqft'] = df_clean['Unit_Size_Sqft'].apply(parse_area)

# Step 3: Convert to float
df_clean['Unit_Size_Sqft'] = df_clean['Unit_Size_Sqft'].astype(float)

# Step 4: Check results
print("Unique count:", df_clean['Unit_Size_Sqft'].nunique())
print("Missing values:", df_clean['Unit_Size_Sqft'].isna().sum())
df_clean[['Unit_Size_Sqft']].head(10)
```

```python
# Check missing values across all columns
df_clean.isnull().sum()
```

```python
import numpy as np

# --------------------------------------------------------
# STEP 1: Check initial missing values
# --------------------------------------------------------
print("Missing values before handling:\n")
print(df_clean.isnull().sum())

# --------------------------------------------------------
# STEP 2: Handle numeric columns (Ticket_Price_Cr, Unit_Size_Sqft, Amenity_Score, etc.)
# Strategy: Replace nulls with the column median 
# --------------------------------------------------------
num_cols = df_clean.select_dtypes(include=['float64','int64']).columns

# Build a dictionary: {column_name : median_value}
num_fill_dict = {col: df_clean[col].median() for col in num_cols}

# Apply fillna once for all numeric columns
df_clean.fillna(num_fill_dict, inplace=True)

# --------------------------------------------------------
# STEP 3: Handle categorical columns (Developer_Name, Booking_Status, Possession_Status, etc.)
# Strategy: Replace nulls with "Unknown"
# --------------------------------------------------------
cat_cols = df_clean.select_dtypes(include=['object']).columns

# Build a dictionary: {column_name : "Unknown"}
cat_fill_dict = {col: "Unknown" for col in cat_cols}

# Apply fillna once for all categorical columns
df_clean.fillna(cat_fill_dict, inplace=True)

# --------------------------------------------------------
# STEP 4: (Optional) Drop columns with >50% null values (if any remain)
# --------------------------------------------------------
threshold = 0.5 * len(df_clean)  # 50% of rows
df_clean.dropna(axis=1, thresh=threshold, inplace=True)

# --------------------------------------------------------
# STEP 5: Verify that nulls are handled
# --------------------------------------------------------
print("\nMissing values after handling:\n")
print(df_clean.isnull().sum())
```

```python
# --------------------------------------------------------
# STEP 1: Price per Sqft (₹ per square foot)
# Ticket_Price_Cr is in Crores → convert to Rupees (1 Cr = 10,000,000)
# Formula: (Price in ₹) / Unit_Size_Sqft
# --------------------------------------------------------
if "Ticket_Price_Cr" in df_clean.columns and "Unit_Size_Sqft" in df_clean.columns:
    df_clean['Price_per_Sqft'] = (
        df_clean['Ticket_Price_Cr'] * 10000000
    ) / df_clean['Unit_Size_Sqft']

# --------------------------------------------------------
# STEP 2: Quarter Number (if we have a booking or sale date column)
# Example: if you have a column like "Booking_Date" or "Sale_Date"
# --------------------------------------------------------
date_col = None
for col in df_clean.columns:
    if "date" in col.lower():   # try to auto-detect
        date_col = col
        break

if date_col:
    # Convert to datetime safely
    df_clean[date_col] = pd.to_datetime(df_clean[date_col], errors='coerce')
    
    # Extract year, month, quarter
    df_clean['Year'] = df_clean[date_col].dt.year
    df_clean['Month'] = df_clean[date_col].dt.month
    df_clean['Quarter_Number'] = df_clean[date_col].dt.quarter
else:
    print("No date column found for Quarter_Number derivation")

# --------------------------------------------------------
# STEP 3: Booking Flag (1 if booked, 0 otherwise)
# Works on Booking_Status or similar column
# --------------------------------------------------------
status_col = None
for col in df_clean.columns:
    if "booking" in col.lower() or "status" in col.lower():
        status_col = col
        break

if status_col:
    df_clean['Booking_Flag'] = df_clean[status_col].apply(
        lambda x: 1 if str(x).strip().lower() in ["booked", "confirmed", "sold"] else 0
    )
else:
    print("No Booking_Status column found for Booking_Flag derivation")

# --------------------------------------------------------
# STEP 4: Quick preview of new columns
# --------------------------------------------------------
df_clean.head(10)
```

```python
# --------------------------------------------------------
# STEP 1: Save cleaned dataframe as a new CSV
# --------------------------------------------------------
output_path = "Luxury_Housing_Bangalore_Cleaned.csv"
df_clean.to_csv(output_path, index=False)

print(f"✅ Cleaned data exported successfully to: {output_path}")
```
